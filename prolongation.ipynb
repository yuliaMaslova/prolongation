## Вспомогательная таблица первого стопа проектов

Создаю дополнительную таблицу `first_stop` сравнивая месяца из которой, будем убирать не валидные проекты при рассчете коэффициента пролонгации. Если `first stop` (первый стоп по проекту) случился после пролонгации, уже впериод второго договора, мы сможем рассчитать коэффициент для менеджера для первого договора, так как он продлил котракт. 
А если первый стоп произошел в период договора раньше его закрытия, то мы не будем учитывать такой проект при подсчете коэффициента пролонгации. Работа по проекту была прекращена и менеджер не проводил пролонгацию.

```python
month_order=columns_fin
stops=alarm[alarm.value.isin(['стоп','end'])].copy()
stops['pos'] = stops['month'].apply(month_order.index)
first_stop = (stops.sort_values('pos').drop_duplicates('id')[['id', 'month']])  # у каждого id берём первую строку
```

## Расчет коэффициента 1 и 2
Чтобы посчитать коэффициент 1 и 2, я написала метод `add_coef`, который рассчитывает коэффициенты 1 и 2, и помещает их в таблицу `finance` как `Месяц_год_коэфициент1`. 
**Для рассчета коэффицентов** не беру ячейки со значением 0, если они не встречаются в таблице alarm. Если же ячейки встречаются в таблице, то для значения stop, end использую созданную таблицу `first_stop` для сравнения: месяц  для рассчета коэффициента идет позже месяца, в котором проект проект поставили на стоп? Если позже, то такой проект не идет в рассчет коэффициента. Если раньше, то такой коэффициент может быть рассчитан.

Также **проверяю значения месяца** с помощью таблицы `zeros`. Если это значение присутсвует, то я сразу меняюю значение на сумму из предыдущего месяца, которая созранена в таблице `zeros`, сохраняю значение суммы как `correct_val`.

Для рассчета коэффициента 1 беру ту часть таблицы, которая удовлетворяет всем условиям:
- не было стопа ранее в период договора
- итоговая **корректная** сумма в месяц окончания договора не NaN и не 0
- сумма в месяце, в котором осуществляется продление не 0 и не NaN

Для рассчета коэффициента 2 аналогично:
- не было стопа ранее в период договора
- итоговая **корректная** сумма в месяц окончания договора не NaN и не 0
- сумма в месяце, в котором осуществляется продление не 0 и не NaN **(и это второй месяц)**
-**сумма в первой месяце, после окончания договора равна 0 или NaN**

```python
columns_fin
```

```python

def add_coef(finance, first_stop, month_prev, month_next, month_next_next, month_order, zeros, prolongation):

    term=prolongation[['id','month']]
    planned = term.loc[term['month'] == month_prev, 'id']
    
    idx_prev = month_order.index(month_prev)                       
    # беру столбец month (строка) из таблицы первого стопа first_stop
    # для каждой строки перевожу его в порядковый номер внутри month_order где хранятся месяца по порядку
    # first_stop['month'].apply(month_order.index)    → 0, 1, 2, …
    #  номер ≤ idx_prev (то есть первый stop у проекта был в базовом месяце или раньше — такие проекты нужно исключить)
    # помещаю эти строки для которых рассчитывать коэф не буду в таблицу плохих индексов bad_idx беря только id
    bad_ids = first_stop.loc[first_stop['month'].apply(month_order.index) <= idx_prev, 'id']

    # вытаскиваю нули нужно месяца
    z = zeros.loc[zeros['month'] == month_prev, ['id', 'prev_sum']]

    # создаю  временную таблицу, из которой буду брать значения, если проект попал в ноль
    base = (finance[['id', month_prev]].merge(z, on='id', how='left'))
    
    #теперь проверяю, если в finance для данного месяца 0, есть ли он в нулях.
    # если есть, то беру значение суммы предыдущего месяца, если нет, то оставляю 0
    correct_val = base.apply(
        lambda r: r['prev_sum'] if (r[month_prev] == 0)
                  else r[month_prev], axis=1 ) 
    
    ok = (
        ~finance['id'].isin(bad_ids)
        & correct_val.notna() & (correct_val != 0)
        & finance[month_next].notna() & (finance[month_next] != 0)
         & finance['id'].isin(planned)
    )
    
    coef_col = f"{month_next}_коэффициент1" 
    finance[coef_col] = pd.NA 
    finance.loc[ok, coef_col] = finance.loc[ok, month_next] / correct_val[ok]

    
    ok2 = (
    ~finance['id'].isin(bad_ids)
    & correct_val.notna() & (correct_val != 0)
    & (finance[month_next].isna() | (finance[month_next] == 0))  #  вот здесь важное уточнение для таблицы 2 ok2
    & finance[month_next_next].notna() & (finance[month_next_next] != 0)
    & finance['id'].isin(planned))
    coef_col2 = f"{month_next_next}_коэффициент2" 
    finance[coef_col2] = pd.NA 
    finance.loc[ok2, coef_col2] = finance.loc[ok2, month_next_next] / correct_val[ok2]
    

for i in range(len(columns_fin)-2):
    month_prev=columns_fin[i]
    month_next=columns_fin[i+1]
    month_next_next=columns_fin[i+2]
    add_coef(finance, first_stop, month_prev, month_next, month_next_next, month_order, zeros, prolongation)
```

## Соединение таблицы с окончанием проектов `prolongation` и коэффициентов пролонгации
Итак, у меня есть таблица с коэффициентами 1, 2 `coefficients`. Теперь остается ее соединить с таблицей `prolongation`
1) Добавляю дополнительные столбцы к таблице `prolongation`, которые называются `next_month` и `next_next_month`. Именно в эти месяцы отвественные аккаунт-менеджеры должны были осуществлять пролонгцацию договоров. Для этого использую схраненный порядок месяцев `month_order`.
2) Добавляю коэффициент 1 и 2 из таблицы с коэффициентами, сверяя нужные месяцы.

```python
prolongation['next_month'] = prolongation['month'].apply(
    lambda m: month_order[month_order.index(m) + 1]
              if month_order.index(m) + 1 < len(month_order)
              else np.nan)

prolongation['next_next_month'] = prolongation['month'].apply(
    lambda m: month_order[month_order.index(m) + 2]
              if month_order.index(m) + 2 < len(month_order)
              else np.nan)
coeff = coefficients.set_index('id')          

prolongation['coef1'] = prolongation.apply(
    lambda r: coeff.loc[r.id, f"{r.next_month}_коэффициент1"]
              if pd.notna(r.next_month)
              and f"{r.next_month}_коэффициент1" in coeff.columns
              else np.nan,
    axis=1)

prolongation['coef2'] = prolongation.apply(
    lambda r: coeff.loc[r.id, f"{r.next_next_month}_коэффициент2"]
              if pd.notna(r.next_next_month)
              and f"{r.next_next_month}_коэффициент2" in coeff.columns
              else np.nan,
    axis=1)
```

## Таблица с менеджером, месяцем и коэффициентами пролонгации
Мне кажется важным добавить количество проектов. Так как менеджер Петрова, например, может иметь высокий коэффициент пролонгации в одном проекте, и в итоге в месяц и в год. Но при этом начать работу позже остальных, или не иметь положительныъ кейсов пролонгации в другие месяцы

```python
prolongation_coef1=prolongation.drop(['month', 'next_next_month', 'coef2'], axis=1).copy()
prolongation_coef1 = prolongation_coef1[prolongation_coef1['coef1'].notna()]
prolongation_coef1=prolongation_coef1.groupby(['AM', 'next_month']).agg(projects = ('id', 'size'), coef1_avg = ('coef1','sum')).reset_index()
prolongation_coef1['coef1_avg']=prolongation_coef1['coef1_avg']/prolongation_coef1['projects']
prolongation_coef1=prolongation_coef1.rename(columns={"next_month": "month", "projects": "projects_coef1"} )
prolongation_coef1['projects_coef2']=pd.NA
prolongation_coef1['coef2_avg']=pd.NA
prolongation_coef1
```

```python
prolongation_coef2=prolongation.drop(['month', 'next_month', 'coef1'], axis=1).copy()
prolongation_coef2 = prolongation_coef2[prolongation_coef2['coef2'].notna()]
prolongation_coef2=prolongation_coef2.groupby(['AM', 'next_next_month']).agg(projects = ('id', 'size'), coef2_avg = ('coef2','sum')).reset_index()
prolongation_coef2['coef2_avg']=prolongation_coef2['coef2_avg']/prolongation_coef2['projects']
prolongation_coef2=prolongation_coef2.rename(columns={"next_next_month": "month", "projects": "projects_coef2"})
prolongation_coef2['projects_coef1']=pd.NA
prolongation_coef2['coef1_avg']=pd.NA
prolongation_coef2=prolongation_coef2.iloc[:, [0, 1 ,4, 5, 2, 3]]
prolongation_coef2
```

```python
final=pd.concat([prolongation_coef1, prolongation_coef2])
final
```

```python
final2=final.groupby(['AM', 'month']).agg(projects_coef1 = ('projects_coef1', 'sum'),
                                           coef1_avg = ('coef1_avg','sum'), 
                                           projects_coef2 = ('projects_coef2', 'sum'), 
                                           coef2_avg = ('coef2_avg','sum')).reset_index()

final2
```

```python
final2['coef1_avg'] = final2['coef1_avg'] / final2['projects_coef1'].replace(0, np.nan)
final2['coef2_avg'] = final2['coef2_avg'] / final2['projects_coef2'].replace(0, np.nan)
final2
```

## Отбор результатов за 2023 год
Можно использовать таблицу `final2` для рассчета коэффициентов за 2024 год тоже. Но в задании, мы опредляем коэффициенты за 2023 год. Поэтому фильтрую итоговую таблицу

```python
valid_months=['Январь 2023',
 'Февраль 2023',
 'Март 2023',
 'Апрель 2023',
 'Май 2023',
 'Июнь 2023',
 'Июль 2023',
 'Август 2023',
 'Сентябрь 2023',
 'Октябрь 2023',
 'Ноябрь 2023',
 'Декабрь 2023']
```

```python
final_2023=final2.loc[final2.month.isin(valid_months)]
final_2023
```

Добавляю пропущенные месяцы, если коэффициентов для менеджера не встречались в опредлеенные месяцы. 
В предыдыдущей таблице `final_2o23` всего 63 строки, а менеджеров 8.

```python
idx = pd.MultiIndex.from_product(
    [final_2023['AM'].unique(), valid_months],
    names=['AM', 'month']
)


final_2023_full = final_2023.set_index(['AM', 'month']).reindex(idx).reset_index()

# Заполняем пропущенные числа проектов нулём (к оставляем NaN)
final_2023_full[['projects_coef1', 'projects_coef2', 'total projects']] = \
    final_2023_full[['projects_coef1', 'projects_coef2', 'total projects']].fillna(0)
final_2023_full
```

## Годовая таблица
Создаю годовую таблицу с результатами.

```python
year = (
    final_2023
    .groupby('AM', as_index=False)
    .agg(
        months_with_prolongation = ('month', 'nunique'),
        projects_coef1=('projects_coef1', 'sum'),
        coef1_num  = ('coef1_avg', lambda s: (s * final_2023.loc[s.index, 'projects_coef1']).sum()),
        projects_coef2=('projects_coef2', 'sum'),
        coef2_num  = ('coef2_avg', lambda s: (s * final_2023.loc[s.index, 'projects_coef2']).sum())
    )
)

year['coef1'] = year['coef1_num'] / year['projects_coef1'].replace(0, np.nan)
year['coef2'] = year['coef2_num'] / year['projects_coef2'].replace(0, np.nan)

year.drop(columns=['coef1_num', 'coef2_num'], inplace=True)
year
```

## Активность менеджеров
Также решила создать таблицу с активностью менеджеров по месяцам. По ней можно предположить и проверить, кто  и когда уходил в отпуск, и почему коэффициентов пролонгации за определенный период нет.
Можно предположить, что большинство уходило в отпуск в июне, а менеджер Петрова начала работать только в конце 2023 года.

```python
temp = final_2023.copy()
temp['active'] = 1
grouped = temp.groupby(['AM', 'month'])['active'].sum().unstack(fill_value=0)

# Упорядочиваем столбцы по valid_months, если какие-то месяцы отсутствуют — добавляем с нулями
for month in valid_months:
    if month not in grouped.columns:
        grouped[month] = 0

grouped = grouped[valid_months]  
activity = grouped.reset_index()
activity
```
